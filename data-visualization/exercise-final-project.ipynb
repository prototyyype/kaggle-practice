# Optional Step: Cleaning Data
I noticed my call references for the columns were coming back with many errors. Using the columns method, I was able to see that there were unneccessary spaces in the column titles. I took this moment as an opportunity to run my data through a hygiene protocol.


### Let's clean up the typos in the column names

## Step 4: Visualize the data

Use the next code cell to create a figure that tells a story behind your dataset.  You can use any chart type (_line chart, bar chart, heatmap, etc_) of your choosing!

````python
# Create a plot
plt.figure(figsize=(25,15))
plt.title("Hip Hop Sample Genres")
sns.countplot(data=my_data, x='year', hue='genre')

# Check that a figure appears below
step_4.check()
````

While detailed, this graph doesn't quite express the story I am looking for. Keep reading to see how I interpolated the data into more descriptive visualizations.

````python
plt.figure(figsize=(25,15))
plt.title("Hip Hop Sample Genres")

sns.histplot(data=my_data, x='year', hue='genre')
````

````python
sns.set_style("dark")
plt.figure(figsize=(25,15))
plt.title("Hip Hop Sample Genres")

sns.kdeplot(data=my_data, x='year', hue='genre')
#, multiple="fill"
````

## Assumptions and Bias

Making a note that this dataset (which is very small with a sum value counts of 144) is already heavily bias with Hip-Hop and R&B/Soul making up 53% and 29%, respectively. 
However, it is interesting to see how certain genres came and went while others continued to be sampled heavily into the 21st century. 

### Some notable mentions...
Initially, R&B/Soul was the most popular genre to be sampled. Following closely, the most popular and consistent genre throughout was Hip-Hop. The two genres intersect around the year 1980 - just following the birth of hiphop in the mid to late 1970s. This explains the soft blue peak in 1970 before the boom in the mid 80's.
